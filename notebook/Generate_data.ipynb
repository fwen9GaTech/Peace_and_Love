```python
import jsonlines
import shutil

# mypath = "deepfill"
# origin_path = "../hateful_memes.zip (Unzipped Files)/hateful_memes/"
file_lst1 = []
file_lst2 = []
with jsonlines.open('train.jsonl') as f:
    i = 0
    for line in f.iter():
        src1 = "deepfill/" + line['img'][4:]
        src2 = "../hateful_memes.zip (Unzipped Files)/hateful_memes/img/" + line['img'][4:]
        dst = "../no_text_70/img/" + line['img'][4:]

        if (line['img'][4:] in deepfill_files) and (random.randint(1, 10) < 8):
          shutil.copyfile(src1, dst)
          file_lst1.append(line['img'][4:])
        else:
          shutil.copyfile(src2, dst)
          file_lst2.append(line['img'][4:])

        i+= 1
        if i %100 ==0:
          print(i)
        #   break
```
